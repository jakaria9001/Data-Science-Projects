```
# Loading the libraries

import pandas as pd #used for data manipulation
from matplotlib import pyplot as plt  # data visualisation
import seaborn as sns   # data visualisation
```

```
# Uploading the dataset to Colab
from google.colab import files
uploaded = files.upload()
```

```
# Loading the IPL matches dataset
ipl_data = pd.read_csv('/content/matches.csv')

# ipl_data
```

```
# First five records
ipl_data.head()
```

```
# No of raws and columns present in the dataset
ipl_data.shape
```

```
# Frequency of players with player of the match award:
ipl_data['player_of_match'].value_counts()
```

```
# Top 5 players with most player of the match award:
ipl_data['player_of_match'].value_counts()[0:5]
```

```
# Bar plot for top 5 players with most man of the matches.
plt.figure(figsize=(8,5)) 
plt.bar(list(ipl_data['player_of_match'].value_counts()[0:5].keys()), list(ipl_data['player_of_match'].value_counts()[0:5]), color='g')
plt.show()
```

```
# Frequency of the result column
ipl_data['result'].value_counts()
```

```
# Frequency & bar graph of the no of toss won w.r.t. each team
ipl_data['toss_winner'].value_counts()
plt.bar(list(ipl_data['toss_winner'].value_counts().keys()), list(ipl_data['toss_winner'].value_counts()), color='b')
plt.xticks(rotation=75)
plt.show()
```

```
# Extracting the records where toss winning team batted first
batting_first_winner = ipl_data[ipl_data['win_by_runs']!=0] 
print(batting_first_winner.head())
```

```
# Histogram for win by run coloumn
plt.figure(figsize=(7,7))
plt.hist(batting_first_winner['win_by_runs'])
plt.title("Distribution of winning runs")
plt.xlabel('Runs ')
plt.ylabel('No of matches')
plt.show()
```

```
# Finding out no of wins for each team after winning the toss
batting_first_winner['winner'].value_counts()
```

```
# Making a bar graph between first three teams winning the match:
plt.figure(figsize=(7,7))
plt.bar(batting_first_winner['winner'].value_counts()[0:3].keys(),batting_first_winner['winner'].value_counts()[0:3], color=["blue","yellow","red"])
plt.xlabel('Team')
plt.ylabel('No of wins')
plt.show()
```

```
# Making a piechart 
plt.figure(figsize=(7,7))
plt.pie(list(batting_first_winner['winner'].value_counts()),labels = list(batting_first_winner['winner'].value_counts().keys()), autopct='%0.1f%%' )
plt.show()
```

```
# Extracting records for which the team won after batting second
winner_batting_second = ipl_data[ipl_data['win_by_wickets']!=0]
winner_batting_second.head()
```

```
# Histogram to see the distribution by how many wickets each team won after betting second
plt.figure(figsize=(7,7))
plt.hist(winner_batting_second['win_by_wickets'],bins=30)
plt.xlabel('Wickets remaining')
plt.ylabel('No of matches')
plt.show()
```

```
# frequency of no wins for teams batting second
winner_batting_second['winner'].value_counts()
```

```
# Bar plot for top 3 champs won betting first
plt.figure(figsize=(7,7))
plt.bar(list(winner_batting_second['winner'].value_counts().keys()[0:3] ), list(winner_batting_second['winner'].value_counts()[0:3]), color=['purple','blue','red'])
plt.xlabel('Team Name')
plt.ylabel('No of matches won')
plt.show()
```

```
# Pie Chart for teams performance when batting second
plt.figure(figsize=(7,7))
plt.pie(list(winner_batting_second['winner'].value_counts()), labels = list(winner_batting_second['winner'].value_counts().keys()))
plt.show()
```
